```python
import pandas as pd
from transformers import BertTokenizer, BertForSequenceClassification, Trainer, TrainingArguments
from datasets import Dataset
import torch
from sklearn.metrics import accuracy_score

df = pd.read_csv("mock_finance_email_dataset.csv")
print("Dataset:")
print(df)

tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')

def tokenize_function(examples):
    return tokenizer(examples["email"], padding="max_length", truncation=True, max_length=512)

dataset = Dataset.from_pandas(df)
tokenized_datasets = dataset.map(tokenize_function, batched=True)

train_test_split = tokenized_datasets.train_test_split(test_size=0.2)
train_dataset = train_test_split["train"]
eval_dataset = train_test_split["test"]

training_args = TrainingArguments(
    output_dir='./results',
    evaluation_strategy="epoch",
    learning_rate=2e-5,
    per_device_train_batch_size=8,
    per_device_eval_batch_size=8,
    num_train_epochs=10,
    weight_decay=0.01,
)

# Carregar o modelo BERT para classificação
model = BertForSequenceClassification.from_pretrained('bert-base-uncased', num_labels=2)

# Definir o objeto Trainer
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=eval_dataset,
    compute_metrics=lambda p: {
        "accuracy": accuracy_score(p.label_ids, torch.argmax(torch.tensor(p.predictions), axis=1).numpy())
    }
)

# Treinar o modelo
print("\nTreinando o modelo...")
trainer.train()

# Avaliar o modelo
metrics = trainer.evaluate()
print(f"\nMetrics: {metrics}")

# Calcular e imprimir a acurácia
predictions = trainer.predict(eval_dataset)
accuracy = accuracy_score(predictions.label_ids, torch.argmax(torch.tensor(predictions.predictions), axis=1).numpy())
print(f"Acurácia: {accuracy}")

model.save_pretrained('/content/email_bert_classifier')
tokenizer.save_pretrained('/content/email_bert_classifier')

# Carregar o modelo salvo para inferência
model = BertForSequenceClassification.from_pretrained('/content/email_bert_classifier')
tokenizer = BertTokenizer.from_pretrained('/content/email_bert_classifier')

# Função para fazer inferências
def classify_email(email_text):
    inputs = tokenizer(email_text, return_tensors='pt', max_length=512, truncation=True, padding='max_length')
    with torch.no_grad():
        outputs = model(**inputs)
    predictions = torch.argmax(outputs.logits, dim=1).item()
    return 'Produtivo' if predictions == 1 else 'Improdutivo'

email_exemplo = "Este é um e-mail importante para a reunião de amanhã."
classificacao = classify_email(email_exemplo)
print(f"\nClassificação: {classificacao}")
```
